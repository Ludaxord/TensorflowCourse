# Matplotlib Basics

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [0, 1, 2]
y = [100, 200, 300]
```

```python
plt.plot(x, y)
plt.show()
```

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3], 'price':[100, 120, 190, 200, 230, 310, 330, 305]})
```

```python
housing
```

```python
plt.scatter(housing['rooms'], housing['price'])
plt.show()
```

```python
plt.plot(x,y, color='#1537d9', marker='o', markersize=15, linestyle='--')
plt.xlim(0, 2)
plt.ylim(0,300)
plt.title("Title")
plt.xlabel('X label')
plt.ylabel('Y label');
```

# Seaborn Basics

```python
import seaborn as sns
```

```python
df = pd.read_csv('DATA/heart.csv')
```

```python
sns.distplot(df['age'])
```

```python
plt.figure(figsize=(8,4))
sns.distplot(df['age'], kde=False, color='red')
```

```python
sns.distplot(df['age'], kde=False, bins=2)
```

```python
df.head()
```

```python
sns.countplot(x='target', data=df)
```

```python
sns.countplot(x='cp', data=df, hue='sex', palette='terrain')
```

```python
sns.boxplot(x='sex', y='age', data=df)
```

```python
sns.boxplot(x='target', y='thalach', data=df, hue='sex')
```

```python
sns.scatterplot(x='chol', y='trestbps', data=df, hue='sex', palette='Dark2', size='age')
```

```python
iris = pd.read_csv('DATA/iris.csv')
```

```python
iris
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```
